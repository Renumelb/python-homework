Create a Python script to analyse financial records of the company to determine the below results.

Pseudocode:
- Create new git repository and required folders and files
- Download budget_data.csv file into the PyBank folder in Jupyter Labs
- Import financial data file budget_data.csv into the Python code file
- Create varaiables:
    - Empty list- hold month & profit/loss. Another to hold net increase/decrease in profit/ loss and month
    - Sum_Total= to hold the total profit/loss over the period. 
    - Sum_Period= to hold the number of months in the period.
    - Maximum and mimimum- month and profit/loss values
- Create counter to read the csv file
- Calculate the total number of months
- Calculate the net profit/loss
- Calculate the greatest increase in profit
- Calculate the greatest decrease in losses
- Calculate the average change in profit/loss
- Print results

Code as per below:

```python
from pathlib import Path
import csv
print(f"Current Working Directory: {Path.cwd()}")
working_file = Path('./budget_data.csv')
with open(working_file, "r") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvfile) 
    # initialise variables
    # make profitloss list contain both date and profit respectively and
    # netchnage list to contain increase/decrease in monthly profit/loss
    profitloss = [] # to hold profit by month#
    netchange = [] # to hold net change by month#
    totbudmonth = 0 # to hold summary profit/loss value #
    totprofitloss = 0 # to hold number of months count #
    line = 0 
# perform calculations - #calculate total profit, total months and net change
    for row in csvreader:
        line += 1
        totprofitloss += int(row[1])
        totbudmonth += 1        
        #print(totbudmonth) to check calculations
        #print(totprofitloss) to check calculations
        # to calculate net change, add months and profit/loss to a dictionary
        profitloss.append([row[0],int(row[1])])
        #print([profitloss])
        if line > 1:
            netchange.append([row[0], int(row[1])-int(profitloss[line-2][1])])
    #print(netchange)    
    # to calculate the average of changes to Profit/Loss
    diff1 = int(profitloss[len(profitloss)-1][1]) - int(profitloss[0][1])
    diff2 = round(diff1/(len(profitloss)-1),2)
    #print(netchange)
    #calculate the maximum increase in profits and maximum decrease in profits
    minimum_value= 0
    for mth, profit in netchange:
        if minimum_value == 0:
            minimum_value = profit
            minimum_key = mth
        elif profit < minimum_value:
            minimum_value = profit
            minimum_key = mth
    #print(minimum_value, minimum_key)
    maximum_value=0
    for mth, profit in netchange:
        if maximum_value == 0:
            maximum_value = profit
            maximum_key = mth
        elif profit > maximum_value:
            maximum_value = profit
            maximum_key = mth
    #print(maximum_value, maximum_key)
    
# Print results
print(f"The total number of months included: {totbudmonth}")
print (f"The net total amount of Profit/Loss over the entire period: ${totprofitloss}")
print(f"The average of the changes in Profit/Loss over the entire period: ${diff2}")
print(f"The greatest increase in profits over the entire period: {maximum_key} ${maximum_value}")
print(f"The greatest decrease in losses is: {minimum_key} ${minimum_value}")
#Set output file name for results
output_path = 'Results.txt'
#Open the output path as a file object
with open(output_path, 'w') as resultsfile:
    # Write daily_average to the output file, convert to string
    resultsfile.write(f"The total number of months included: {totbudmonth}\n")
    resultsfile.write(f"The net total amount of Profit/Loss over the entire period: ${totprofitloss}\n")
    resultsfile.write(f"The average of the changes in Profit/Loss over the entire period: ${diff2}\n")
    resultsfile.write(f"The greatest increase in profits over the entire period: {maximum_key} ${maximum_value}\n")
    resultsfile.write(f"The greatest decrease in losses is: {minimum_key} ${minimum_value}.")
```
